```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calculate_values(t, A0, k1, k2):
    A = A0 * np.exp(-k1 * t)
    B = (k1 * A0 / (k1 - k2)) * (np.exp(-k2 * t) - np.exp(-k1 * t))
    C = A0 + (A0 / (k1 - k2)) * (k2 * np.exp(-k1 * t) - k1 * np.exp(-k2 * t))
    return pd.Series({
        'A': np.round(A, 3),
        'B': np.round(B, 3),
        'C': np.round(C, 3)
    })

A0 = 1
k1 = 0.30
k2 = 0.15
I = 21
J = 53
t_val = [i for i in range(I)]

kinetics = pd.DataFrame([calculate_values(t, A0, k1, k2) for t in t_val])
print(kinetics)
```

```python
plt.xlabel('time')
plt.ylabel('C true')
plt.title('Kinetics')
plt.legend(frameon=False)
plt.plot(kinetics['A'], color='red', label='A')
plt.plot(kinetics['B'], color='blue', label='B')
plt.plot(kinetics['C'], color='green', label='C')
plt.show()
```

```python
def errors_generator(I, J):
  errors = np.zeros((I, J))
  dRMSE = 1.5
  for i in range(I):
    for j in range(J):
      dA = -6
      for k in range(12):
        dA = dA + np.random.random()
      errors[i, j] = dA * dRMSE * 0.01
      yield errors[i, j]

error_generator = errors_generator(I, J)
RMSE = pd.DataFrame(np.fromiter(error_generator, float).reshape(I, J))
RMSE
```

```python
x = kinetics.dot(transposed_spectra) + RMSE
dataDf = pd.DataFrame(x)
dataDf
```

```python
a = [1]
b = [0.0, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5,
     0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5]
c = [0.00, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75,
     0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75]
window = pd.DataFrame({
'A': 1,
'B': b,
'C': c
})
window['t'] = range(21)
window.set_index('t', inplace=True)
window
```

```python
# матрица с_in
data = {
    'A': [1.005, 0.771, 0.638, 0.503, 0.411, 0.314, 0.255, 0.210, 0.155, 0.137,
      0.098, 0.079, 0.065, 0.059, 0.035, 0.027, 0.014, 0.017, -0.001, 0.000, -0.002],
    'B': [0.002, 0.170, 0.272, 0.326, 0.326, 0.331, 0.312, 0.288, 0.249, 0.218,
      0.188, 0.162, 0.137, 0.118, 0.086, 0.074, 0.044, 0.032, 0.013, -0.006, -0.003],
    'C': [-0.004, 0.047, 0.096, 0.173, 0.268, 0.358, 0.435, 0.492, 0.595, 0.646,
      0.708, 0.765, 0.787, 0.825, 0.865, 0.893, 0.956, 0.953, 0.997, 1.010, 0.999]
    }

с_in = pd.DataFrame(data)
с_in
```

```python
last_c_out = None
last_c_hat = None
last_s_hat = None
for _ in range(10):
  c_adj = с_in.copy()
  c_adj = с_in.where((с_in >= 0) & (window != 0), 0)
  c_adj.loc[0, 'A'] = 1
  closure = c_adj.sum(axis=1)
  c_hat = c_adj.div(closure, axis=0)
  c_hat = c_hat.round(3)
  c_hat.columns = ['Ahat', 'Bhat', 'Chat']
  transposed_c_hat = c_hat.transpose()
  reversed_c_hat = np.linalg.inv(transposed_c_hat.dot(c_hat))
  reversed_c_hat = reversed_c_hat.round(3)
  transposed_data = dataDf.transpose()
  s_in = transposed_data.dot(c_hat.dot(reversed_c_hat))
  s_hat = s_in.copy()
  s_hat[s_hat < 0] = 0
  s_hat = s_hat.round(3)
  s_hat.columns = ['Ahat', 'Bhat', 'Chat']
  transposed_s_hat = s_hat.transpose()
  reversed_s_hat = np.linalg.inv(transposed_s_hat.dot(s_hat))
  reversed_s_hat = reversed_s_hat.round(3)
  c_out = dataDf.dot(s_hat.dot(reversed_s_hat))
  c_out = c_out.round(3)
  c_in = c_out
  last_c_out = c_out
  last_c_hat = c_hat
  last_s_hat = s_hat

matrixes_c = pd.concat([kinetics, last_c_hat], axis=1)
matrixes_c
```

```python
# матрица с_in
data = {
    'A': [1.005, 0.715, 0.547, 0.395, 0.303, 0.205, 0.152, 0.116, 0.075, 0.067,
          0.038, 0.028, 0.023, 0.023, 0.010, 0.007, 0.004, 0.011, 0.000, 0.008, 0.005],
    'B': [-0.001, 0.246, 0.397, 0.483, 0.493, 0.510, 0.491, 0.463, 0.418, 0.379,
          0.343, 0.312, 0.278, 0.255, 0.214, 0.199, 0.163, 0.145, 0.123, 0.096, 0.099],
    'C': [-0.001, 0.028, 0.062, 0.125, 0.210, 0.289, 0.359, 0.412, 0.507, 0.555,
          0.613, 0.667, 0.689, 0.724, 0.762, 0.789, 0.848, 0.846, 0.887, 0.900, 0.890]
    }
с_in = pd.DataFrame(data)
с_in
```

```python
last_c_out = None
last_c_fit = None
last_s_fit = None
Rate1 = 0.311
Rate2 = 0.1485
values = np.arange(0, 21)
for _ in range(300):
  c_hat = с_in.copy()
  c_hat = с_in.where((с_in >= 0) & (window != 0), 0)
  A_fit = np.exp(-Rate1 * values)
  rounded_a_fit = np.round(A_fit, 3)
  B_fit = Rate1 / (Rate1 - Rate2) * (np.exp(-Rate2 * values) -
                                     np.exp(-Rate1 * values))
  rounded_b_fit = np.round(B_fit, 3)
  C_fit = 1 + 1/(Rate1 - Rate2) * (Rate2 * np.exp(-Rate1 * values) -
                                   Rate1 * np.exp(-Rate2 * values))
  rounded_c_fit = np.round(C_fit, 3)
  data = {
  'Afit': rounded_a_fit,
  'Bfit': rounded_b_fit,
  'Cfit': rounded_c_fit
  }
  с_fit = pd.DataFrame(data)
  transposed_с_fit = с_fit.transpose()
  reversed_с_fit = np.linalg.inv(transposed_с_fit.dot(с_fit))
  reversed_с_fit = reversed_с_fit.round(3)
  transposed_data = dataDf.transpose()
  s_in = transposed_data.dot(с_fit.dot(reversed_с_fit))
  s_fit = s_in.copy()
  s_fit[s_fit < 0] = 0
  s_fit = s_fit.round(3)
  s_fit.columns = ['Afit', 'Bfit', 'Cfit']
  transposed_s_fit = s_fit.transpose()
  reversed_s_fit = np.linalg.inv(transposed_s_fit.dot(s_fit))
  reversed_s_fit = reversed_s_fit.round(3)
  c_out = dataDf.dot(s_fit.dot(reversed_s_fit))
  c_out = c_out.round(3)
  c_in = c_out
  last_c_out = c_out
  last_c_fit = с_fit
  last_s_fit = s_fit

matrixes_c = pd.concat([kinetics, last_c_hat, last_c_fit], axis=1)
matrixes_c
```
